# Emotion Weighting System Documentation

---

## Problem

Our app captures emotions in two different ways:
1. **In-App**: When users actively share their feelings inside the app
2. **Background**: Continuous emotion tracking while using their phone normally


---

## Solution

We give **more weight to in-app emotions** because the user is consciously sharing their experience. But we also check background emotions to make sure they're being honest (sometimes people hide their true feelings).

**Our Basic Rule:**
```
Final Emotion = (65% × In-App) + (35% × Background)
```

### Why 65-35?

**In-App gets 65% because:**
- User is thinking about their feelings right now
- They're sharing specific experiences
- It's directly related to what they want to talk about

**Background gets 35% because:**
- Shows their real day-to-day emotions
- Catches things they might not report
- Helps verify if they're being truthful

---

## Workflow

```
┌──────────────────┐
│ User opens app   │
│ Shares feelings  │ ───────► In-App Data (65%)
└──────────────────┘              │
                                  │
┌──────────────────┐              ▼
│ Phone tracks     │         ┌─────────────┐
│ emotions daily   │ ──────► │   Combine   │──► Final Result
└──────────────────┘         └─────────────┘
Background Data (35%)             
```

---

## When We Adjust the Weights

Sometimes 65-35 isn't perfect. We adjust based on these situations:

### 1. How Long They Used the App

- **Quick use (< 2 min)**: Maybe they're rushed → 55% in-app, 45% background
- **Normal use (2-10 min)**: Standard → 65% in-app, 35% background  
- **Deep session (> 10 min)**: They're really engaged → 75% in-app, 25% background

**Why?** Longer time = more reliable data from the app.

### 2. Do Both Match?

- **Emotions match**: They say "happy" and background shows happy → 70% in-app, 30% background
- **Somewhat different**: Small differences → Keep 65-35
- **Big mismatch**: They say "happy" but background shows "sad" → 50-50 split

**Why?** When there's a mismatch, we need to investigate both equally.

---

## Examples

### Example 1: Everything Matches 
Sarah opens the app after a great day at work. She types "I'm feeling really happy today!"

- **In-App**: Happy (80%)
- **Background**: Mostly happy emotions all day (75%)
- **They match!** So we trust the in-app more → 70% in-app, 30% background
- **Final Result**: Happy (78%) - High confidence

### Example 2: Something's Off 
John says "I'm doing great!" in the app, but background tracking shows stress and anxiety all day.

- **In-App**: Happy (90%)
- **Background**: Stressed (70%)
- **Big mismatch!** We need to look at both equally → 50% in-app, 50% background
- **Final Result**: Mixed emotions (60% happy, 35% stressed) - Flag for review

This might mean John is hiding his feelings or needs support.


---

## Special Cases to Handle

**1. No Background Data Yet**
- New user or they disabled background tracking
- Use 100% in-app (we only have that data)

**2. User Barely Used the App**
- Opened app, clicked one button, closed it
- Use more background (60%) since in-app isn't reliable

**3. Very First Time User**
- Start with standard 65-35
- Learn their patterns over time

**4. Emergency Situations**
- If we detect severe depression or harmful thoughts
- Alert immediately, regardless of weights

---

## Privacy & Ethics

We're tracking emotions, so privacy matters:

1. **Tell users clearly** that we track both in-app and background
2. **Let them opt out** of background tracking anytime
3. **Keep background data only 30 days** then delete it
4. **Show them the weights** so they understand how we calculated their results
